```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler

# Load the dataset
data = pd.read_csv('test.csv')

# Define features (X) and target variable (y)
X = data.drop('EI', axis=1)  # Features
y = data['EI'].astype(int)    # Target variable (convert to integer if necessary)

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Create and train the MLP model
mlp = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=1000, activation='relu', random_state=42)
mlp.fit(X_train, y_train)

# Make predictions
predictions = mlp.predict(X_test)

# Evaluate the model
conf_matrix = confusion_matrix(y_test, predictions)
class_report = classification_report(y_test, predictions, zero_division=0)

# Output results
print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)

# Additional output: Model configuration
print("Model Configuration:")
print(f"Activation Function: {mlp.activation}")
print(f"Hidden Layer Sizes: {mlp.hidden_layer_sizes}")
print(f"Max Iterations: {mlp.max_iter}")

```
